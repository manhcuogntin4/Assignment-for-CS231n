```python
Hs={}
```

```python
for i in xrange(len(Hidden_layer_sizes)):
    X=D if i==0 else Hs[i-1]
    fan_in=X.shape[1]
    fan_out=Hidden_layer_sizes[i]
    W=np.random.randn(fan_in,fan_out)*0.01
    H=np.dot(X,W)
    H=act[nonlinears[i]](H)
    #print(act[nonlinears[i]](H))
    Hs[i]=H
    #print X[:10]
```
